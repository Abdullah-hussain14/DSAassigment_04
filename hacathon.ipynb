```python
def city_col(cols):
    if cols == 1:
        return 'Karachi'
    elif cols == 2:
        return 'Lahore'
    elif cols == 3:
        return 'Islamabad'
    elif cols == 4:
        return 'Faisalabad'
    elif cols == 5:
        return 'Quetta'
    elif cols == 6:
        return 'Peshawar'
    else:
        return cols
```

```python
df['City']= df['City'].apply(city_col)
df['City']
```

```python
df.columns
```

```python
df['Order Priority']
```

```python
code_map={
    0.0 : 'LOW',
    1.0 : 'MEDIUM',
    2.0 : 'HIGH',
    3.0 : 'CRITICAL'
}
df['Order Priority']=df['Order Priority'].map(code_map)
df['Order Priority']
```

```python
df7 = df.pivot_table(values='total', index='Month', columns='category', aggfunc='sum')

df7.plot(kind='bar', figsize=(12,6))
plt.title("Monthly Sales by Category")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=25)
plt.show()
```

### Sales across all categories increased from January to March. Electronics showed a sharp rise, while Furniture had the slowest growth
